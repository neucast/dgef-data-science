```python
# ********* Ejercicio 1 *********
# Escribir un programa que permita gestionar la base de datos de clientes de una empresa. Los clientes 
# se guardarán en un diccionario en el que la clave de cada cliente será su RFC, y el valor será otro 
# diccionario con los datos del cliente (nombre, dirección, teléfono, correo, preferente), donde preferente 
# tendrá el valor True si se trata de un cliente preferente. El programa debe preguntar al usuario por 
# una opción del siguiente menú:

# (1) Añadir cliente, 
# (2) Eliminar cliente, 
# (3) Mostrar cliente, 
# (4) Listar todos los clientes, 
# (5) Listar clientes preferentes, 
# (6) Terminar.

# En función de la opción elegida el programa tendrá que hacer lo siguiente:
# Preguntar los datos del cliente, crear un diccionario con los datos y añadirlo a la base de datos. Preguntar 
# por el RFC del cliente y eliminar sus datos de la base de datos. Preguntar por el RFC del cliente y mostrar 
# sus datos. Mostrar lista de todos los clientes de la base datos con su RFC y nombre. Mostrar la lista de 
# clientes preferentes de la base de datos con su RFC y nombre. Terminar el programa.

# Agrega cliente 1
cliente = {"nombre":"Roberto Pérez", "direccion":"Avenida 5 de mayo 2", 
          "telefono":"55-5432-1234", "correo":"abc@a.com", "preferente":"SI"}
datos   = {'ROBE1234': cliente }
# Agrega cliente 2
cliente = {"nombre":"Emilio Pérez", "direccion":"Tepepan, _mirador", 
          "telefono":"55-9999-8888", "correo":"emi@a.com", "preferente":"SI"}
datos.update({'EMIL123456': cliente})
# Agrega cliente 3
cliente = {"nombre":"Lalo Quera", "direccion":"Psicologos 35", 
          "telefono":"55-2345-1111", "correo":"lalo@locos.com", "preferente":"NO"}
datos.update({'LALO123456': cliente})

def imprimeCliente( clienteDiccionario ):
    print("Nombre: " + clienteDiccionario["nombre"] + \
          ", dirección: " + clienteDiccionario["direccion"]+ \
          ", teléfono: " + clienteDiccionario["telefono"]+ \
          ", correo: " + clienteDiccionario["correo"]+ \
          ", preferente: " + clienteDiccionario["preferente"])
 
opcion = 0
while opcion != 6:
    print("************* MENÚ DE OPCIONES ********* ") 
    print("(1) Añadir cliente") 
    print("(2) Eliminar cliente") 
    print("(3) Mostrar cliente") 
    print("(4) Listar todos los clientes")  
    print("(5) Listar clientes preferentes") 
    print("(6) Terminar") 
    
    #Pide que el Usuario ingrese una opción
    opcion = int(input('\nEscriba una opción del menú: '))
    
    if opcion == 1:# Añadir cliente
        print( "_______________________________________ Opción 1, Añadir cliente")
        RFC        = input('\nRFC: ') 
        nombre     = input('Nombre: ') 
        direccion  = input('Dirección: ') 
        Telefono   = input('Teléfono: ') 
        correo     = input('Correo: ') 
        preferente = input('Es preferente, SI/NO: ') 
        cliente = {"nombre":nombre, "direccion":direccion,  "telefono":Telefono, 
                  "correo":correo, "preferente":preferente.upper()}
        datos.update({RFC: cliente})
        print( "Se dió de alta al cliente: " + RFC )
        print( cliente )
        
    elif opcion == 2: # Eliminar cliente
        print( "_______________________________________ Opción 2, Eliminar cliente")
        RFC =input('RFC: ')
        datos.pop(RFC)
        print( "Se eliminó al cliente:" + RFC )
        
    elif opcion == 3: # Mostrar cliente
        print( "_______________________________________ Opción 3, Mostrar cliente")
        RFC = ""
        RFC = input('RFC: ')
        cliente = datos[RFC]
        print( "Los datos del Cliente " + RFC + " son:")
        imprimeCliente(cliente)
    
    elif opcion == 4: # Listar todos los clientes
        print( "_______________________________________ Opción 4, Listar todos los clientes")
        for RFC in datos:
            cliente = datos[RFC]
            print("Datos del Cliente " + RFC +":")
            imprimeCliente( cliente )
            
    elif opcion == 5: # Listar clientes preferentes
        print( "_______________________________________ Opción 5, Listar clientes preferentes")
        encontroDatos = False
        for RFC in datos:
            cliente = datos[RFC]
            if cliente["preferente"] == 'SI':
                encontroDatos = True
                print("Datos del Cliente preferente " + RFC +":")
                imprimeCliente(cliente)
        if encontroDatos == False:
            print("No existen clientes preferentes")
            
print("Fin del programa")
```

```python
# ********* Ejercicio 5 *********
# Escribe un generador de contraseñas en Python. 
# Las contraseñas seguras tienen una combinación de letras minúsculas, letras mayúsculas, números y símbolos. 
# Las contraseñas deben ser aleatorias, generando una nueva contraseña cada vez que el usuario solicite una 
# nueva contraseña. 
# Incluya su código en una funcion.

from random import random

def obtenerPasword(longitud):
    b = ""
    for i in range(0, longitud):
        tipoDeCaracter = int(random()*4)
        if tipoDeCaracter == 0 :
            a = chr(int(65+random()*( 90-65)) )# Maysculas: A,B,C,...,Z
        elif tipoDeCaracter == 1 :
            a = chr(int(97+random()*(122-97)) )# Minusculas: a,b,c,...,z
        elif tipoDeCaracter == 2 :
            a = chr(int(48+random()*(57-48)) )# Números: 0,1,2,...,9
        else :
            a = chr(int(33+random()*(38-33)) )# Carácteres especiales: !"#$%&
        b = b + a # Acumula las letras de la contraseña
    return b
    
longitudDelPasword = int(input('\nEscriba la longitud de la contraseña: '))

print( "La contraseña sugerida es: " + obtenerPasword( longitudDelPasword ) )
```

```python
# ********* Ejercicio 6 *********
# Utilizando el archivo "Historico_Vector_M.csv", calcular el rendimiento del Precio Limpio 
# en un horizonte temporal de dos días y eliminar los renglones con errores.

import pandas as pd
import numpy as np
#Importa el CSV en un dataframe
df = pd.read_csv("/Users/roberto/Local/Phyton/CURSO/Ejericios_Modulo2_Parte1/Historico_Vector_M.csv", 
                 parse_dates=[1])
#Ordena los datos por fecha
df.sort_values(by=['FECHA'])
df.head()
#Calula el promedio del "PRECIO SUCIO" para todas las fechas del archivo CSV
tabla = pd.pivot_table(df, index =["FECHA"], aggfunc={"PRECIO LIMPIO":np.mean})
#Genera un campo adicional "PRECIO SUCIO anterior dos días" con el precio sucio de dos días anteriores a la fecha
tabla["PRECIO LIMPIO anterior dos días"] = tabla["PRECIO LIMPIO"].shift(-2)
#Genera un campo adicional dónde cálcula el rendimiento en por ciento: PRECIO SUCIO(n-1)/PRECIO SUCIO(n) *100 
tabla["Rendimiento"] = (tabla["PRECIO LIMPIO anterior dos días"]/tabla["PRECIO LIMPIO"]-1)*100
#Filtra todos los registros que no sean factibles de calcular elimienando los valores "NaN"
tabla = tabla.query('Rendimiento != "NaN"')
#Muestra los resultados
tabla
```

```python
# ********* Ejercicio 8 ********* 

# Paso 1. Definir una función en Python que corresponda a la expresión analítica f(x,y) = x^2 - y^2 + 1. 
# Paso 2. Definir una función en Python que corresponda al gradiente de la función f(x,y). Donde el gradiente 
# se define como Gradiente(f)(x,y) = (fx(x,y),fy(x,y)), con fx la derivada parcial de la función f respecto a 
# "x" y fy la derivada parcial de la función f respecto a "y". 
# Paso 3. Definir una tercer función en Python que, dado un punto (x0,y0) en R^2, evalue el gradiente definido 
# en el Paso 2, y proporcione como resultado el valor de la función f (definida en el Paso 1) evaluada en el 
# punto (x1,y1) = (x0,y0) - 0.5Gradiente(f)(x,y). 
# Paso 4. Si en el Paso 3, (x0,y0) = (1,1), ¿Cuál relación de orden es correcta, 
# f(x1,y1) > f(x0,y0) o f(x1,y1) < f(x0,y0)?

x= 1
y= 1

#Paso 1, Función de la expresión f(x,y) = x^2 - y^2 + 1
def F(x,y):
    return (x*x) - (y*y) +1
    
#Paso 2, Obtención de una función correspondiente al gradiente de la función, Gradiente(f)(x,y) = (fx(x,y),fy(x,y))
def GradienteDeF(x,y):
    return [2*x,-2*y]
#Paso 3, evalue el gradiente (x1,y1) = (x0,y0) - 0.5Gradiente(f)(x,y)
def EvaluaGradiente(x,y):
    [x1,y1] = GradienteDeF(x,y)
    return [x-0.5*x1,y-0.5*y1]

print("Paso 3, Evaluacón del gradiente: ", EvaluaGradiente(x,y))
r1 = F(x,y)
r2 = EvaluaGradiente(x,y)
if  r1 > pow(r2[0]*r2[0]+r2[1]*r2[1], 0.5): 
    print("Paso 4, Resultado: f(x1,y1) > f(x0,y0)")
else:
    print("Paso 4, Resultado: f(x1,y1) < f(x0,y0)")         
```
