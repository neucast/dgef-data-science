### Ejercicio 1

Escribir un programa que permita gestionar la base de datos de clientes de una empresa. Los clientes se guardarán en un diccionario en el que la clave de cada cliente será su RFC, y el valor será otro diccionario con los datos del cliente (nombre, dirección, teléfono, correo, preferente), donde preferente tendrá el valor True si se trata de un cliente preferente. El programa debe preguntar al usuario por una opción del siguiente menú: 

(1) Añadir cliente, 
<br>(2) Eliminar cliente, 
<br>(3) Mostrar cliente, 
<br>(4) Listar todos los clientes, 
<br>(5) Listar clientes preferentes, 
<br>(6) Terminar. 

En función de la opción elegida el programa tendrá que hacer lo siguiente:

Preguntar los datos del cliente, crear un diccionario con los datos y añadirlo a la base de datos.
Preguntar por el RFC del cliente y eliminar sus datos de la base de datos.
Preguntar por el RFC del cliente y mostrar sus datos.
Mostrar lista de todos los clientes de la base datos con su RFC y nombre.
Mostrar la lista de clientes preferentes de la base de datos con su RFC y nombre.
Terminar el programa.

#### Solución propuesta por Gabriel Vázquez

```python
#Función validaEntero
#Valida que una cadena corresponda a un valor entero
def validarEntero(cadena, rangoInferior = -2147483648, rangoSuperior = 2147483648):
    #Mensajes de error
    diccErrores = {1000 : "Sin errores",                                        \
                   1001 : "El valor ingresado no es numérico",                  \
                   1101 : "El valor ingresado está fuera del rango definido",   \
                   9999 : "Error no clasificado"}
    #Variables locales
    banderaError = True
    mensajeError = diccErrores[1000]
    valor        = 0

    #Revisa que "cadena" sea un valor entero
    if cadena.isdigit():
        valor = int(cadena)
        #Revisa que "cadena" esté dentro del rango
        if not(rangoInferior <= valor and valor <= rangoSuperior):
            banderaError = False
            mensajeError = diccErrores[1101]
    else:
        banderaError = False
        mensajeError = diccErrores[1001]
    
    #Regresa resultados
    return not(banderaError), mensajeError, valor


#Función mostrarMenu
#Muestra el menú para gestionar la base de clientes
def mostrarMenu():
    #Opciones del menú
    opciones = ["(1) Añadir cliente",              \
                "(2) Eliminar cliente",            \
                "(3) Mostrar cliente",             \
                "(4) Listar todos los clientes",   \
                "(5) Listar clientes preferentes", \
                "(6) Terminar"]

    #Muestra el menú
    print("\nGestión de clientes\n")
    for li in range(len(opciones)):
        print("    " + opciones[li])
    print()


#Función validaREGEX
#Valida que la cadena corresponda a una expresión regular válida
def validarREGEX(cadena, regex):
    #Librerías
    import re
    #Mensajes de error
    diccErrores = {1000 : "Sin errores",                        \
                   2000 : "La expresión regular no coincide",   \
                   9999 : "Error no clasificado"}
    #Variables locales
    banderaError = True
    mensajeError = diccErrores[1000]
    patron       = re.compile(regex)
    nCadena      = cadena.strip()

    #Busca el patrón
    encontrado = patron.match(nCadena)
    if encontrado:
        if len(nCadena) == len(encontrado.group()):
            banderaError = False
        else:
            mensajeError = diccErrores[2000]
            nCadena = encontrado.group()
    else:
        mensajeError = diccErrores[2000]

    #Regresa resultados
    return not(banderaError), mensajeError, nCadena


#Función solicitarRFC
#Solicita y valida un RFC
def solicitarRFC(diccionario):
    #Variables locales
    regexRFC  = "^([A-ZÑ]{3,4}([0-9]{2})(0[1-9]|1[0-2])(0[1-9]|1[0-9]|2[0-9]|3[0-1]))([-])?([A-Z0-9]{2}[0-9A])?$"
    rfc       = ""
    existeRFC = False
    
    #Solicita RFC
    while True:
        rfc = input("RFC del cliente: ")
        validador = validarREGEX(rfc.upper(), regexRFC)
        if not(validador[0]):
            print("RFC inválido, los formatos válidos son XAXX010101A00, XAX010101A00")
        else:
            #Verifica la existencia del RFC en el diccionario
            if validador[2] in diccionario:
                print("RFC existente en el diccionario")
                existeRFC = True
                rfc = validador[2]
            else:
                rfc = validador[2]
            break

    #Regresa resultados
    return existeRFC, rfc

                
#Funcion agregarCliente
#Agrega un cliente
def agregarCliente(diccionario):
    #Variables locales
    regexTel        = "^([2-9]{1}[0-9]{9}$)"
    regexCorreo     = "^[(a-z0-9\_\-\.)]+@[(a-z0-9\_\-\.)]+\.[(a-z)]{2,15}$"
    regexPreferente = "^([SN]{1})"
    banderaError    = True
    
    #Pregunta datos
    print()
    print("Introduce un nuevo cliente")
    #RFC
    rfc = solicitarRFC(diccionario)[1]
    #Resto de los datos
    while banderaError:
        banderaError = False
        nombre       = input("Nombre completo: ")
        direccion    = input("Dirección completa: ")
        telefono     = input("Teléfono (10 dígitos): ")
        correo       = input("Correo electrónico: ")
        preferente   = input("Cliente preferente (S/N): ")

        #Valida los datos introducidos
        #Teléfono
        validador = validarREGEX(telefono, regexTel)
        if not(validador[0]):
            banderaError = True
            print("Teléfono inválido, el formato es a 10 dígitos ej. 2299338844")
        else:
            telefono = validador[2]
        #Correo electrónico
        validador = validarREGEX(correo.lower(), regexCorreo)
        if not(validador[0]):
            banderaError = True
            print("Correo inválido, el formato válido es correo_electronico@dominio.ext")
        else:
            correo = validador[2]
        #Preferente
        validador = validarREGEX(preferente.upper(), regexPreferente)
        if not(validador[0]):
            banderaError = True
            print("Opción inválida, seleccione S-Preferente y N-No preferente")
        else:
            if validador[2] == 'S':
                preferente = True
            else:
                preferente = False

    #Crea un diccionario con la información del cliente
    elemento = dict(nombre = nombre, direccion = direccion, telefono = telefono, correo = correo, preferente = preferente)
    #Agrega el nuevo elemento
    diccionario[rfc] = elemento


#Funcion eliminarCliente
#Elimina un cliente existente usando su RFC
def eliminarCliente(diccionario, rfc):
    #Variables locales
    banderaError = True

    #Revisa si existe el cliente antes de eliminarlo 
    if rfc in diccionario:
        del diccionario[rfc]
        banderaError = False
    
    return not(banderaError)


#Funcion imprimirDetalleCliente
#Imprime en pantalla todo el detalle de un cliente usando su RFC
def imprimirDetalleCliente(rfc, cliente):
    print()
    print("RFC: " + rfc)
    print("Nombre completo:",    cliente['nombre'])
    print("Dirección:",          cliente['direccion'])
    print("Teléfono:",           cliente['telefono'])
    print("Correo electrónico:", cliente['correo'])
    print("Cliente preferente:", cliente['preferente'])

    
#Funcion imprimirCliente
#Imprime en pantalla un cliente usando su RFC
def imprimirCliente(rfc, cliente):
    print()
    print("RFC: " + rfc)
    print("Nombre completo:",    cliente['nombre'])

    
#Funcion mostrarClientes
#Imprime en pantalla los clientes
def mostrarClientes(diccionario, rfc = ''):
    if len(rfc) > 0:
        imprimirDetalleCliente(rfc, diccionario[rfc])
    else:
        for llave, valor in diccionario.items():
            imprimirCliente(llave, valor)

            
#Funcion mostrarClientesPreferentes
#Imprime en pantalla los clientes preferentes
def mostrarClientesPreferentes(diccionario):
    for llave, valor in diccionario.items():
        if valor['preferente']:
            imprimirCliente(llave, valor)


#Función baseDatosClientes
#Manejador de la base de datos de clientes
def baseDatosClientes(diccionario, opcion):
    #Agrega un cliente
    if opcion == 1:
        agregarCliente(diccionario)
    #Elimina un cliente
    elif opcion == 2:
        salidaRFC = solicitarRFC(diccionario)
        #Solo si el cliente existe
        if salidaRFC[0]:
            eliminarCliente(diccionario, salidaRFC[1])
            print("Cliente eliminado:", salidaRFC[1])
        else:
            print("No existe el cliente:", salidaRFC[1])
    #Muestra un cliente
    elif opcion == 3:
        salidaRFC = solicitarRFC(diccionario)
        #Solo si el cliente existe
        if salidaRFC[0]:
            mostrarClientes(diccionario, salidaRFC[1])
        else:
            print("No existe el cliente:", salidaRFC[1])
    #Muestra todos los clientes
    elif opcion == 4:
        print("\nListado de clientes")
        mostrarClientes(diccionario)
    #Muestra todos los clientes preferentes
    elif opcion == 5:
        print("\nListado de clientes preferentes")
        mostrarClientesPreferentes(diccionario)


#Función ejercicio_1
#Programa principal ejercicio 1
def ejercicio_1():
    #Variables locales
    clientesDiccionario = {}
    
    #Captura una elección
    while True:
        mostrarMenu()
        opcion = input("Por favor, selecciona una opción válida del menú: ")
        regresoFunc = validarEntero(opcion, 1, 6)
        #Opción inválida o válida
        if regresoFunc[2] >= 1 and regresoFunc[2] <= 5:
            baseDatosClientes(clientesDiccionario, regresoFunc[2])
        elif regresoFunc[2] == 6:
            print("Fin del programa")
            break
        else:
            print("Opción no válida:", regresoFunc[1])

            
#Ejecución programa principal
ejercicio_1()
```

```python
#Función validaREGEX
#Valida que la cadena corresponda a una expresión regular válida
def validarREGEX(cadena, regex):
    #Librerías
    import re
    #Mensajes de error
    diccErrores = {1000 : "Sin errores",                        \
                   2000 : "La expresión regular no coincide",   \
                   9999 : "Error no clasificado"}
    #Variables locales
    banderaError = True
    mensajeError = diccErrores[1000]
    patron       = re.compile(regex)
    nCadena      = cadena.strip()

    #Busca el patrón
    encontrado = patron.match(nCadena)
    if encontrado:
        if len(nCadena) == len(encontrado.group()):
            banderaError = False
        else:
            mensajeError = diccErrores[2000]
            nCadena = encontrado.group()
    else:
        mensajeError = diccErrores[2000]

    #Regresa resultados
    return not(banderaError), mensajeError, nCadena


#Funcion binarioADecimal
#Convierte una cadena binaria a valor decimal
def binarioADecimal(cadena):
    #Variables locales
    valor       = 0
    maxPotencia = len(cadena) - 1
    
    for i in range(len(cadena)):
        if cadena[i] == '1':
            valor = valor + pow(2, maxPotencia)
        maxPotencia -= 1
    
    return valor


#Función conversionBinariaADecimal
#Solicita el ingreso de una cadena en formato binario para convertirla en decimal
def conversionBinariaADecimal():
    #Variables locales
    regexBinario    = "^([01]{1,60}$)"
    cadena          = ""
    valor           = 0
    
    while True:
        cadena = input("Ingresa una cadena en binario (0,1): ")
        validador = validarREGEX(cadena, regexBinario)
        if not(validador[0]):
            banderaError = True
            print("Cadena binaria inválida (valores permitidos 0,1), o cadena muy larga")
        else:
            valor = binarioADecimal(validador[2])
            print("La cadena binaria <", validador[2], "> equivale en decimal a <", valor, ">")
            break


#Funcion decimalABinario
#Convierte una cadena decimal a valor binario
def decimalABinario(cadena):
    #Variables locales
    valor       = int(cadena)
    valorStr    = ""
    cadenaBin   = []
    
    # Convierte a binario
    while valor > 0:
        cadenaBin.extend([valor % 2])
        valor //= 2
    #Ordena la cadena
    for i in range(len(cadenaBin)):
        valorStr = valorStr + str(cadenaBin[len(cadenaBin)-i-1])
        
    return valorStr


#Función conversionDecimalABinaria
#Solicita el ingreso de una cadena en formato decimal para convertirla en binario
def conversionDecimalABinaria():
    #Variables locales
    regexDecimal    = "^([0-9]{1,14}$)"
    cadena          = ""
    valor           = "0"
    
    while True:
        cadena = input("Ingresa una cadena en decimal (0-9): ")
        validador = validarREGEX(cadena, regexDecimal)
        if not(validador[0]):
            banderaError = True
            print("Cadena decimal inválida (valores permitidos 0-9), o cadena muy larga")
        else:
            valor = decimalABinario(validador[2])
            print("La cadena decimal <", validador[2], "> equivale en binario a <", valor, ">")
            break


#Función ejercicio_2
#Programa principal ejercicio 2
def ejercicio_2():
    print("\nConversión de binario a decimal")
    conversionBinariaADecimal()
    print("\nConversión de decimal a binario")
    conversionDecimalABinaria()


#Ejecución programa principal
ejercicio_2() 
```

### Ejercicio 5

Escribe un generador de contraseñas en Python. 
<br>Las contraseñas seguras tienen una combinación de letras minúsculas, letras mayúsculas, números y símbolos. 
<br>Las contraseñas deben ser aleatorias, generando una nueva contraseña cada vez que el usuario solicite una nueva contraseña. <br>Incluya su código en una funcion.

#### Solución propuesta por Luis Newmann

```python

import string
import random

LETTERS = string.ascii_letters
NUMBERS = string.digits
SYMBOLS = string.punctuation


# Inner class to statically store terminal colors.
class BColors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'


"""
Asks for the length of the password.
:returns number <class 'int'>
"""


def get_password_length():
    length = input("How long do you want your password? ")
    return int(length)


"""
Generates a random password having the specified length.
:length -> length of password to be generated. Defaults to 8 chars if nothing is specified.
:passwordCombinationChoiceList-> a list of boolean values representing a user choice for
    string constant to be used to generate password.
    0 item ---> digits
         True to add digits to constant False otherwise.
    1 item ---> letters
         True to add letters to constant False otherwise.
    2 item ---> symbols
         True to add symbols to constant False otherwise.
:returns string <class 'str'>
"""


def password_generator(passwordCombinationChoiceList, length=16):
    # create alphanumerical by fetching string constant
    printable = fetch_string_constant(passwordCombinationChoiceList)

    # convert printable from string to list and shuffle
    printable = list(printable)
    random.shuffle(printable)

    # generate random password
    random_password = random.choices(printable, k=length)

    # convert generated password to string
    random_password = "".join(random_password)
    return random_password


"""
Returns a string constant based on users passwordCombinationChoiceList.
string constant can either be digits, letters, symbols or
combination of them.
: passwordCombinationChoiceList --> list <class 'list'> of boolean
    0 item ---> digits
        True to add digits to constant False otherwise
    1 item ---> letters
        True to add letters to constant False otherwise
    2 item ---> symbols
        True to add symbols to constant False otherwise
"""


def fetch_string_constant(passwordCombinationChoiceList):
    string_constant = ""

    string_constant += NUMBERS if passwordCombinationChoiceList[0] else ""
    string_constant += LETTERS if passwordCombinationChoiceList[1] else ""
    string_constant += SYMBOLS if passwordCombinationChoiceList[2] else ""

    return string_constant


"""
Sets the agreement of the secure combination of characters for a password.
:returns list <class 'list'> of boolean [True, True, True]
    0 item ---> digits
    1 item ---> letters
    2 item ---> symbols
"""


def secure_password_combination_choice():
    return [True, True, True]


"""
Prompt a user to choose password character combination which
could either be digits, letters, symbols or combination of
either of them.
:returns list <class 'list'> of boolean [True, True, False]
    0 item ---> digits
    1 item ---> letters
    2 item ---> symbols
"""


def password_combination_choice():
    # Retrieve a user's password character combination choice.
    wantDigits = input("Want digits  (Yes/No)? ").upper() == "YES"
    wantLetters = input("Want letters  (Yes/No)? ").upper() == "YES"
    wantSymbols = input("Want symbols (Yes/No)? ").upper() == "YES"

    return [wantDigits, wantLetters, wantSymbols]


if __name__ == '__main__':

    wantsAnotherPassword = True

    while wantsAnotherPassword:
        length = get_password_length()
        passwordCombinationChoiceList = secure_password_combination_choice()
        password = password_generator(passwordCombinationChoiceList, length)

        print(BColors.OKBLUE + "The generated password is:" + BColors.ENDC, password)

        wantsAnotherPassword = input("Do you want to generate another password (Yes/No)? ").upper() == "YES"
```

<br> Paso 3. Definir una tercer función en Python que, dado un punto (x0,y0) en R^2,  evalue  el gradiente definido en el Paso 2, y proporcione como resultado el valor de la función f (definida en el Paso 1) evaluada en el punto (x1,y1) = (x0,y0) - 0.5Gradiente(f)(x,y).
<br> Paso 4. Si en el Paso 3, (x0,y0) = (1,1), ¿Cuál relación de orden es correcta, f(x1,y1) > f(x0,y0) o f(x1,y1) < f(x0,y0)?

#### Solución propuesta por Ernesto Agonizantes

```python
def f(x,y):
    z = x**2 -y**2 +1
    return(z)

def gradiente(x,y):
    g = (2*x,-2*y)
    return(g)

def f3(x,y):
    a = gradiente(x,y)
    x1 = x-0.5*a[0]
    y1 = y-0.5*a[1]
    b = f(x1,y1)
    return(b)

f1 = f3(1,1)
f0 = f(1,1)
if f0 > f1:
    print("La relación es: f(x1,y1) < f(x0,y0)")
else:
     print("La relación es: f(x1,y1) > f(x0,y0)")  
```

Crear una función lambda que pueda realizar las funciones de suma, resta, división y multiplicación indicadas como parámetros entre 2 números dados.

#### Solución propuesta por Luis Newmann

```python
from math import nan
import numpy as np

"""
Lambda function that can perform the addition, subtraction, division and multiplication 
functions indicated as parameters between two given numbers.
"""


def execute_all_basic_math_operations(value1, value2):
    sum = lambda x, y: x + y
    subtraction = lambda x, y: x - y
    multiplication = lambda x, y: x * y
    division = lambda x, y: x / y if y != 0 else nan
    execute_operation = lambda f, value1, value2: f(value1, value2)

    sumResult = execute_operation(sum, value1, value2)
    subtractionResult = execute_operation(subtraction, value1, value2)
    multiplicationResult = execute_operation(multiplication, value1, value2)
    divisionResult = execute_operation(division, value1, value2)
    return np.array([sumResult, subtractionResult, multiplicationResult, divisionResult])


def get_result(operation, resultSet):
    result = lambda n, resultSet: resultSet[n]

    if operation == "+":
        return result(0, resultSet)
    elif operation == "-":
        return result(1, resultSet)
    elif operation == "*":
        return result(2, resultSet)
    else:
        return result(3, resultSet)


def show_result(value1, value2, operation, resultSet):
    print(value1, operation, value2, "=", get_result(operation, resultSet))


def show_sum_result(value1, value2, resultSet):
    show_result(value1, value2, "+", resultSet)


def show_subtraction_result(value1, value2, resultSet):
    show_result(value1, value2, "-", resultSet)


def show_multiplication_result(value1, value2, resultSet):
    show_result(value1, value2, "*", resultSet)


def show_division_result(value1, value2, resultSet):
    show_result(value1, value2, "/", resultSet)


def show_all_basic_math_operations_results(value1, value2):
    # Compute results.
    resultSet = execute_all_basic_math_operations(value1, value2)

    # Show results.
    show_sum_result(value1, value2, resultSet)  # sum result.
    show_subtraction_result(value1, value2, resultSet)  # subtraction result.
    show_multiplication_result(value1, value2, resultSet)  # multiplication result.
    show_division_result(value1, value2, resultSet)  # division result.
    
show_all_basic_math_operations_results(6,0)
```
