### Ejercicio. Utilizando el dataset CAPITALIZACION_RComputo.csv, generar una tabla que contenga el Capital Neto Vigente para cada institución y Fecha y genere el total sistema

```python
import pandas as pd
```

```python
datos = pd.read_csv("/Users/cecilia/Documents/BM/Curso DS/CAPITALIZACION_RComputo.csv",dtype='str',encoding = "ISO-8859-1")
```

```python
datos
```

```python
datos["ICAP"] = datos["ICAP"].astype(float)
datos["Capital Neto Vigente"] = datos["Capital Neto Vigente"].astype(float)
datos["Capital Neto reg. vig. dic 2012"] = datos["Capital Neto reg. vig. dic 2012"].astype(float)
datos["Operaciones Sujetas a Riesgo"] = datos["Operaciones Sujetas a Riesgo"].astype(float)
datos["RC Total"] = datos["RC Total"].astype(float)
datos["RC x riesgo de Credito"] = datos["RC x riesgo de Credito"].astype(float)
datos["RC x riesgo de Mercado"] = datos["RC x riesgo de Mercado"].astype(float)
datos["RC x riesgo Operacional"] = datos["RC x riesgo Operacional"].astype(float)
datos["RC x riesgo de Filiales del Exterior"] = datos["RC x riesgo de Filiales del Exterior"].astype(float)
```

```python
pd.pivot_table(datos,columns='Institucion',values=['Capital Neto Vigente'], \
                           index = 'Fecha',aggfunc = 'sum', fill_value = 0, \
                           margins = True, margins_name = 'Total Sistema')
```

```python
pd.pivot_table(datos,columns='Fecha',values=['Capital Neto Vigente'], \
                           index = 'Institucion',aggfunc = 'sum', fill_value = 0, \
                           margins = True, margins_name = 'Total Sistema')
```

### Ejercicio. Utilizando el dataset CAPITALIZACION_RComputo.csv, generar una tabla que contenga el monto maximo de los tipos de riesgo sobre el Requerimiento de Capital Total para cada institución y genere el total sistema

```python
datos['RC xr C monto'] = datos['RC x riesgo de Credito'] * datos['RC Total'] / 100
datos['RC xr M monto'] = datos['RC x riesgo de Mercado'] * datos['RC Total'] / 100
datos['RC xr O monto'] = datos['RC x riesgo Operacional'] * datos['RC Total'] / 100
datos['RC xr F monto'] = datos['RC x riesgo de Filiales del Exterior'] * datos['RC Total'] / 100
```

```python
datos
```

```python
pd.pivot_table(datos,index='Institucion',values=['RC xr C monto','RC xr M monto','RC xr O monto','RC xr F monto'], \
                           aggfunc = 'max', fill_value = 0,margins = True, margins_name = 'Total Sistema')
```
