# La librería Pandas proporciona estructuras y métodos para manipular datos

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

# Existen principalmente dos estructuras para manipular datos en Pandas: "Series" y "DataFrame"

# 1. Series

```python
s0 = pd.Series([18.56,20.07,14.05,0.17])
s0
```

```python
#Se puede acceder a los atributos de este objeto
s0.index
```

```python
#Se puede acceder a los atributos de este objeto
s0.values
```

```python
s0.index = ["Peso-Dolar","Peso-Euro","Peso-DolarCanadiense","Peso-YenJapones"]
s0.name = "Tipo de Cambio"
s0
```

```python
#Se puede definir de manera completa
s = pd.Series([18.56, 20.07, 14.05,  0.17],name = "Tipo de Cambio",index = ["M1","M2","M3","M4"])
s
```

```python
#Se puede acceder a sus elementos como un diccionario (A)
s["M1"]
```

```python
#Se puede acceder a sus elementos como un diccionario (B)
s[["M1","M3"]]
```

```python
#Cálculo de estadísticas descriptivas
s.median(), s.mean(), s.std()
```

```python
s.min(), s.max()
```

```python
s.describe()
```

```python
#Graficación rápida en Pandas
fig, axes = plt.subplots(1, 4, figsize=(12, 3))
s.plot(ax=axes[0],kind="line",title="linea")
s.plot(ax=axes[1], kind='bar', title='barra')
s.plot(ax=axes[2], kind='box', title='box-plot')
s.plot(ax=axes[3], kind='pie', title='pay')
```

# 2. DataFrame

```python
#Asignación de índice
MA.set_index("FECHA",inplace=True)
MA.head()
```

```python
#Operaciones con columnas (Error)
MA["SumaPrecios"] = MA["PRECIO SUCIO"] + MA["PRECIO LIMPIO"]
MA.head()
```

```python
#Operaciones con columnas (ajuste de tipo de columna)
MA["PRECIO SUCIO"] = MA["PRECIO SUCIO"].astype(float)
MA["PRECIO LIMPIO"] = MA["PRECIO LIMPIO"].astype(float)
MA["SumaPrecios"] = MA["PRECIO SUCIO"] + MA["PRECIO LIMPIO"]
MA.head()
```

```python
#Eliminar el índice definido anteriormente
MA.reset_index(drop=True,inplace=True)
MA.head()
```

```python
#Ajuste de columna de fecha. Notar como se crea un objeto "datetime"
MA = pd.read_csv("C:\\Users\\ricardo\\Desktop\\BanxicoTrabajoCasa_v3\\CursoCienciaDeDatos\\Clases\\Historico_Vector_M.csv",dtype='str',encoding = "ISO-8859-1")
MA["FECHA"] = pd.to_datetime(MA["FECHA"],dayfirst=True)
MA.head()
```

```python
#Ordenación por fecha (A)
MA.sort_values("FECHA",ascending=True,inplace=True)
MA.head()
```

```python
#Ordenación por fecha (B)
MA.sort_values("FECHA",ascending=False,inplace=True)
MA.head()
```

```python
#Rezago de renglones
MA["PRECIO SUCIO R1"] = MA["PRECIO SUCIO"].shift(-1)
MA.head()
```

```python
#Se genera una celda vacía al final
MA.tail()
```

```python
#Correción rezago (A)
MA.fillna(0,inplace=True)
MA.tail()
```

```python
#Correción rezago (B)
MA.drop(MA.tail(1).index,inplace=True)
MA.tail()
```

```python
#Correción enumeración índice
MA.reset_index(drop=True,inplace=True)
MA.head()
```
